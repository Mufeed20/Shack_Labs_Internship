# **Sentence Transormers**

Let us use Sentence Transformers to find the words that have equal semantic meaning 

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
transformer = SentenceTransformer('all-MiniLM-L6-v2')

amz_embedded = transformer.encode(amz_data.product_name,convert_to_tensor=True)
flp_embedded = transformer.encode(flp_data.product_name,convert_to_tensor=True)

# Cosine Similarities 
cos_score = util.cos_sim(amz_embedded, flp_embedded)
```

```python
cos_score[0]
```

```python
cos_score[1]
```

```python
#let us store the index of a flipkart product that is similar to an amazon product
similar_prod_indices = []

for i in range(20000):
    max_score_idxs = torch.topk(cos_score[i],k=5,largest=True,sorted=True).indices
    for idx in max_score_idxs:
        if idx != i:
            similar_prod_indices.append(idx)
            break
```

```python
# for x in similar_prod_indices:
#   print(x.item())
#   break
```

```python
similar_prod_indices = [id.item() for id in similar_prod_indices]
```

```python
amz_data.product_name
```

```python
flp_prod_data = []
amz_prod_data = []

for id, prod_name in enumerate(amz_data.product_name):
    flp_prod_data.append(flp_data.iloc[similar_prod_indices[id]])
    amz_prod_data.append(amz_data.iloc[id])
```

```python
amz_prod_data
```

```python
flp_prod_data = pd.DataFrame(flp_prod_data)
flp_prod_data.columns = ['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart']
amz_prod_data = pd.DataFrame(amz_prod_data)
amz_prod_data.columns = ['Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon']
```

```python
amz_prod_data.shape, flp_prod_data.shape
```

```python
flipkart_final = flp_prod_data.copy()
flipkart_final = flipkart_final.reset_index(drop=True)
flipkart_final.head()
```

```python
combined_data = pd.concat([flipkart_final,amz_prod_data],axis=1)
combined_data.to_csv('final_result.csv')
```
